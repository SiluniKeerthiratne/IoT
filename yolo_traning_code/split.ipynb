```python
import zipfile
import os

def unzip(zip_file=None, extract_to="./extracted"):
    try:
        # Create the target folder if it does not exist
        if not os.path.exists(extract_to):
            os.makedirs(extract_to)

        # Extract all files to the specified folder
        with zipfile.ZipFile(zip_file, 'r') as z:
            z.extractall(extract_to)
            print(f"Extracted all to {extract_to}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Call the function, passing the zip file path and folder to extract into
unzip('/content/drive/MyDrive/Siluni_Docs/MyData_1.zip', '../content')
```

```python
import os
import shutil
import random

# Paths to original dataset
train_images_dir = "/content/MyData_1/images"
train_labels_dir = "/content/MyData_1/labels"

# Paths to new train/validation dataset
train_images_output_dir = "/content/dataset/train/images"
train_labels_output_dir = "/content/dataset/train/labels"
val_images_output_dir = "/content/dataset/val/images"
val_labels_output_dir = "/content/dataset/val/labels"

# Create output directories if they don't exist
os.makedirs(train_images_output_dir, exist_ok=True)
os.makedirs(train_labels_output_dir, exist_ok=True)
os.makedirs(val_images_output_dir, exist_ok=True)
os.makedirs(val_labels_output_dir, exist_ok=True)

# Get list of image and label files
image_files = sorted(os.listdir(train_images_dir))
label_files = sorted(os.listdir(train_labels_dir))

# Ensure matching number of images and labels
assert len(image_files) == len(label_files), "Mismatch in image and label file counts."

# Pair images with corresponding labels
file_pairs = list(zip(image_files, label_files))

# Shuffle the data for randomness
random.shuffle(file_pairs)

# Split the dataset
split_idx = int(0.8 * len(file_pairs))
train_files = file_pairs[:split_idx]
val_files = file_pairs[split_idx:]

# Function to copy files to the destination folder
def copy_files(file_pairs, src_image_dir, src_label_dir, dest_image_dir, dest_label_dir):
    for image_file, label_file in file_pairs:
        shutil.copy(os.path.join(src_image_dir, image_file), os.path.join(dest_image_dir, image_file))
        shutil.copy(os.path.join(src_label_dir, label_file), os.path.join(dest_label_dir, label_file))

# Copy training files
copy_files(train_files, train_images_dir, train_labels_dir, train_images_output_dir, train_labels_output_dir)

# Copy validation files
copy_files(val_files, train_images_dir, train_labels_dir, val_images_output_dir, val_labels_output_dir)

print("Dataset split completed!")
```

```python
import shutil
import os

# Path to the dataset directory
dataset_dir = "/content/dataset"

# Output zip file path
dataset_zip_path = "/content/drive/MyDrive/Siluni_Docs/MyData_1_split.zip"

# Zip the entire dataset folder
shutil.make_archive(dataset_zip_path.replace(".zip", ""), 'zip', dataset_dir)

print("Zipping the entire dataset completed!")
```
